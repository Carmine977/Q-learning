## Solution Space of Tower of Hanoi (4 disks)

```python
# Import libraries
import itertools
import numpy as np
import pickle
```

### States Representation  

```python
#4 disks in a tower
p = list(itertools.permutations(['SMLE','*','*'], 3))
comb1 = []
for i in p:
    if i not in comb1:
        comb1.append(i)

## 1 disk in a tower, 3 disks togheter: 
# S disk in a tower and 3 disks togheter
comb2_1 = list(itertools.permutations(['S','MLE','*'], 3))
# M disk in a tower and 3 disks togheter
comb2_2 = list(itertools.permutations(['M','SLE','*'], 3))
# L disk in a tower and 3 disks togheter
comb2_3 = list(itertools.permutations(['L','SME','*'], 3))
# E disk in a tower and 3 disks togheter
comb2_4 = list(itertools.permutations(['E','SML','*'], 3))


## 1 disk in a tower, 1 disk in a tower, 2 disks togheter: 
# S disk in a tower, M disk in a tower and 2 disks togheter
comb3_1 = list(itertools.permutations(['S','M','LE'], 3))
# S disk in a tower, L disk in a tower and 2 disks togheter
comb3_2 = list(itertools.permutations(['S','L','ME'], 3))
# S disk in a tower, E disk in a tower and 2 disks togheter
comb3_3 = list(itertools.permutations(['S','E','ML'], 3))
# M disk in a tower, L disk in a tower and 2 disks togheter
comb3_4 = list(itertools.permutations(['M','L','SE'], 3))
# M disk in a tower, E disk in a tower and 2 disks togheter
comb3_5 = list(itertools.permutations(['M','E','SL'], 3))
# L disk in a tower, E disk in a tower and 2 disks togheter
comb3_6 = list(itertools.permutations(['L','E','SM'], 3))

## 2 disks in a tower, 2 disks in a tower:
# S,M disks in a tower, L,E disks in a tower
comb4_1 = list(itertools.permutations(['SM','LE','*'], 3))
# S,L disks in a tower, M,E disks in a tower
comb4_2 = list(itertools.permutations(['SL','ME','*'], 3))
# S,E disks in a tower, M,L disks in a tower
comb4_3 = list(itertools.permutations(['SE','ML','*'], 3))
# M,L disks in a tower, S,E disks in a tower
comb4_4 = list(itertools.permutations(['ML','SE','*'], 3))
# M,E disks in a tower, S,L disks in a tower
comb4_5 = list(itertools.permutations(['ME','SL','*'], 3))
# L,E disks in a tower, S,M disks in a tower
comb4_6 = list(itertools.permutations(['LE','SM','*'], 3))

# Merge of all combinations
comb = comb1+ \
        comb2_1+comb2_2+comb2_3+comb2_4+ \
        comb3_1+comb3_2+comb3_3+comb3_4+comb3_5+comb3_6+ \
        comb4_1+comb4_2+comb4_3+comb4_4+comb4_5+comb4_6
print "Total number of states:", len(comb)
```

### Transitions Between States

```python
def check_position(s, state):
    # Return True if the disk is placed on top of a bigger disk. 
    # Return False if the disk is placed on top of a smaller disk.
    m = {k: v for v, k in enumerate(['S', 'M', 'L', 'E'])}
    state_n = []
    
    if state != '*':
        s_n = m[s[0]]
        for i in state:
            state_n = state_n + [m[i[0]]]
        #print s_n, state_n
        return all(i > s_n for i in state_n)
    else:
        return True
```

```python
def all_combination(state):
    # Return for each state all possible moves according to game rules.    
    # Example: all_combination(('SML', '*', '*')) -> [('ML', 'S', '*'), ('ML', '*', 'S')].
    state_all = []
    #print state
    for tow in range(3):
        ix = [0,1,2]
        group = state[tow]    
        if group != '*':
            ix.remove(tow)
            out = ['','','']
    
            for i in ix:
                out = []
                ind = ix[:]
                mov = group[0]
                rem = group[1:len(group)]
                if rem == '':
                    rem ='*'

                if mov != '*':
                    if check_position(mov, state[i]):
                        out = ['','','']
                        out[tow] = rem
                        if state[i]=='*':
                            out[i] = mov
                        else:
                            out[i] = mov+state[i]
                        ind.remove(i) 
                        out[ind[0]] = state[ind[0]]
            
                if len(out) != 0:
                    state_all = state_all + [(out[0],out[1],out[2])]

    return state_all
```

** Solution space: **  
* All links toward the goal have high values **10**;
* Allowed states transitions have value **0**;
* Not-allowed states transitions have value **-1**.
